```python
import numpy as np
from scipy.stats import poisson, norm
import matplotlib.pyplot as plt
```

```python
mu_list = [2, 10, 20]
sigma = np.sqrt(mu_list)

x = np.arange(-5, 40)

fig, axs = plt.subplots(len(mu_list))
fig.suptitle('Poisson v.s. Normal')
fig.set_size_inches(10, 10)

for index, mu in enumerate(mu_list):
    dist_poisson = poisson(mu)
    dist_normal = norm(loc=mu, scale=sigma[index])
    axs[index].plot(x, dist_poisson.pmf(x), label=f'Poisson $\mu={mu}$')
    axs[index].plot(x, dist_normal.pdf(x), label=f'Normal $\mu={mu}$ $\sigma={sigma[index]:.2f}$')

for ax in axs.flat:
    ax.legend()
    ax.label_outer()

plt.xlabel('x')
plt.show()
```
